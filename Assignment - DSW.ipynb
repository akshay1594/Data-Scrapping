```python
import pandas as pd
import numpy as np
from selenium import webdriver
```

```python
browser = webdriver.Chrome('chromedriver.exe')
```

```python
url='https://www.goodreads.com/choiceawards/best-books-2017'
browser.get(url)
parents=browser.find_element_by_css_selector('div.categoryContainer')
cat=parents.find_elements_by_css_selector('div.clearFix')
len(cat)
```

```python
df=pd.DataFrame()

for books in cat:
    try:
        category=books.find_element_by_css_selector('h4.category__copy').text
    except:
        category=None
    try:
        namep=books.find_element_by_css_selector('div.category__winnerImageContainer')
        name=namep.find_element_by_css_selector('img.category__winnerImage').get_attribute('alt')
    except:
        namep=None
        name=None
    try:
        img1=books.find_element_by_css_selector('div.category__winnerImageContainer')
        img=books.find_element_by_css_selector('img.category__winnerImage').get_attribute('src')
    except:
        img1=None
        img=None
    try:
        description=books.find_element_by_css_selector('a').get_attribute('href')
    except:
        description=None
    
    book_dict=pd.Series({
        'Category':category,'Book Name':name,
        'URL of the Book Image':img1,'Url book desc':description
    })
    
    
    df=df.append(book_dict,ignore_index=True)
```

# For extracting every website

```python
parent12=browser.find_elements_by_css_selector('div.clearFix')
urls=[]

for web in parent12:
    try:
        website=web.find_element_by_css_selector('a').get_attribute('href')
    except:
        website=None
    urls.append(website)
    print(urls)
```

```python
urls
```

```python
book_db=pd.DataFrame()
for i in urls:
    browser.get(i)
    

    parents=browser.find_element_by_css_selector('div.miniBook')
    parents1=browser.find_element_by_css_selector('div.pollContents')
    try:
        authorname=parents.find_element_by_css_selector('span[itemprop="name"]').text
    except:
        authorname=None
    try:
        desc=parents.find_element_by_css_selector('div.readable.stacked.gcaBookDescription').text
    except:
        desc=None
    try:
        winnervotes=parents.find_element_by_css_selector('span.greyText.gcaNumVotes').text
        winnervotes=int(winnervotes.split(" ")[0].replace(",",""))
    except:
        winnervotes=None
    try:
        totvotes=parents1.find_element_by_css_selector('div.greyText.gcaNumVotes').text
        totvotes1=int(totvotes.split(" ")[0].replace(",",""))
    except:
        totvotes1=None

    pervote=(winnervotes/totvotes1)*100
    
    dict=pd.Series({
        'Author Name':authorname,'Description':desc,
        'Winner votes':winnervotes,'Total votes':totvotes1,
        'Percentage':pervote
    })
    book_db=book_db.append(dict,ignore_index=True)
book_db 
```
